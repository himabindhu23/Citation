```python
# Filter the rated restaurants
res_rated = df[df['rating'] != 'Not given'].copy()

# Convert rating column from object to integer
res_rated['rating'] = res_rated['rating'].astype('int')

# Create a dataframe that contains the restaurant names with their rating counts
res_rating_count = res_rated.groupby(['restaurant_name'])['rating'].count().sort_values(ascending = False).reset_index()
res_rating_count.head()
```

```python
# Group by restaurant_name and calculate rating count and average rating
restaurant_stats = res_rated.groupby('restaurant_name')['rating'].agg(['count', 'mean'])
```

```python
# Filter restaurants based on criteria
eligible_restaurants = restaurant_stats[(restaurant_stats['count'] > 50) & (restaurant_stats['mean'] > 4)]

# List of eligible restaurants
eligible_restaurant_names = eligible_restaurants.index.tolist()

print("Restaurants eligible for the promotional offer:")
print(eligible_restaurant_names)
```

```python
rev_above_20 = df[df['cost_of_the_order'] > 20]['cost_of_the_order'] * 0.25

# Calculate revenue for orders between 5 and 20 dollars (charged at 15%)
rev_between_5_and_20 = df[(df['cost_of_the_order'] > 5) & (df['cost_of_the_order'] <= 20)]['cost_of_the_order'] * 0.15

# Calculate total revenue
total_rev = rev_above_20.sum() + rev_between_5_and_20.sum()

print(f"The net revenue generated by the company across all orders is ${total_rev:.2f}.")
```

```python
# Calculate total time (preparation time + delivery time) for each order
df['total_time'] = df['food_preparation_time'] + df['delivery_time']

# Filter orders that take more than 60 minutes
long_delivery_orders = df[df['total_time'] > 60]

# Calculate the percentage
percentage_long_delivery = (len(long_delivery_orders) / len(df)) * 100

print(f"The percentage of orders taking more than 60 minutes for delivery is approximately {percentage_long_delivery:.2f}%.")
```

```python
# Create a new column 'is_weekend' to indicate if the order was placed on a weekend
df['is_weekend'] = df['day_of_the_week'] == 'Weekend'

# Group by 'is_weekend' and calculate the mean delivery time
mean_delivery_time_by_day = df.groupby('is_weekend')['delivery_time'].mean()

# Print the results
print("Mean Delivery Time:")
print(f"On Weekdays: {mean_delivery_time_by_day[False]:.2f} minutes")
print(f"On Weekends: {mean_delivery_time_by_day[True]:.2f} minutes")
```
